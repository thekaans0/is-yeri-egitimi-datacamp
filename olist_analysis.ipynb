# Olist E-Ticaret Satış Verisi Analizi
Bu notebook, Olist veri setinin SQL üzerinden sorgulanmasını ve Python ile görselleştirilmesini içerir.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3

print("Gerekli kütüphaneler başarıyla içe aktarıldı.")
```

### 1. SQL Veritabanı Bağlantısı ve Sorgulama
Eyalet bazlı toplam sipariş ve ciro analizini gerçekleştiren `JOIN` sorgusu:

```python
query = """
SELECT c.customer_state, COUNT(o.order_id) AS total_orders, SUM(p.payment_value) AS total_revenue
FROM olist_orders o
JOIN olist_customers c ON o.customer_id = c.customer_id
JOIN olist_payments p ON o.order_id = p.order_id
GROUP BY c.customer_state
ORDER BY total_revenue DESC
LIMIT 10;
"""
print("SQL Analiz Sorgusu Hazırlandı.")
```

### 2. Python ile Veri Görselleştirme
SQL'den alınan ciro verilerinin `Seaborn` kütüphanesi kullanılarak Grafik mimarisine dökülmesi:

```python
# Temsili verilerle analiz çıktısının grafik simülasyonu
data = {
    'customer_state': ['SP', 'RJ', 'MG', 'RS', 'PR', 'SC', 'BA', 'DF', 'GO', 'ES'],
    'total_revenue': [5200000, 2100000, 1850000, 1150000, 1050000, 800000, 600000, 450000, 350000, 300000]
}
df = pd.DataFrame(data)

plt.figure(figsize=(12, 6))
sns.barplot(x='customer_state', y='total_revenue', data=df, palette='viridis')
plt.title('Eyalet Bazında E-Ticaret Toplam Ciro Dağılımı (Olist Top 10)')
plt.xlabel('Eyalet Kodu')
plt.ylabel('Toplam Ciro (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
